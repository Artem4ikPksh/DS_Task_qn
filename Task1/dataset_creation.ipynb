# Dataset Creation for Mountain NER

This notebook documents the creation of a synthetic BIO-annotated dataset for mountain name recognition.

## Goal
Generate 5000 sentences containing mountain names and annotate them with BIO labels.

```python
mountains=["Mount Everest","K2","Kangchenjunga","Lhotse","Makalu","Cho Oyu","Dhaulagiri","Manaslu","Nanga Parbat","Annapurna","Mount Elbrus","Mont Blanc","Matterhorn","Aconcagua","Denali","Mount Fuji","Mount Cook","Mount Kenya","Mount Kilimanjaro","Ben Nevis"]
```

```python
templates=["I climbed {m} last summer .","The expedition reached {m} yesterday .","Many tourists visit {m} every year .","Snow covered {m} overnight .","We photographed {m} at sunrise ."]
```

```python
def bio(sentence,mountains):
    toks=sentence.split();out=[];i=0
    mts=sorted(mountains,key=lambda x:-len(x.split()))
    while i<len(toks):
        ok=False
        for m in mts:
            ms=m.split()
            if toks[i:i+len(ms)]==ms:
                for j,w in enumerate(ms):
                    out.append((w,'B-MOUNTAIN' if j==0 else 'I-MOUNTAIN'))
                i+=len(ms);ok=True;break
        if not ok:
            out.append((toks[i],'O'));i+=1
    return out
```

```python
import random
sentences=[random.choice(templates).format(m=random.choice(mountains)) for _ in range(5000)]
train,valid,test=sentences[:4000],sentences[4000:4500],sentences[4500:]
print(len(train),len(valid),len(test))
```

```python
def save_conll(data,path):
    with open(path,'w',encoding='utf8') as f:
        for s in data:
            for tok,tag in bio(s,mountains):
                f.write(f'{tok} {tag}\n')
            f.write('\n')
# save_conll(train,'train.txt')
```

The dataset is stored in CoNLL format (`token label`) with labels `O`, `B-MOUNTAIN`, and `I-MOUNTAIN`. It is then split into training, validation, and test subsets.
