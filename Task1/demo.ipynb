```python
import spacy
from spacy import displacy
from transformers import pipeline
```

```python
# Path to your saved model
model_path = "./best_mountain_ner_model"

print("Loading the trained model...")
ner_pipeline = pipeline(
    "ner", 
    model=model_path, 
    tokenizer=model_path,
    aggregation_strategy="first" # Glues subtokens together (e.g. hoverla)
)
print("The model is ready for testing!")
```

```python
def visualize_ner(text):
   # Get predictions from our BERT model
    entities = ner_pipeline(text)
    
    # Format the output data for the SpaCy visualizer
    spacy_ents = []
    for ent in entities:
        spacy_ents.append({
            "start": ent["start"],
            "end": ent["end"],
            "label": "MOUNTAIN"
        })
    
    # Create a document structure dictionary
    ex = [{
        "text": text,
        "ents": spacy_ents,
        "title": None
    }]
    
    # Set the highlight color for the MOUNTAIN label (e.g. bright green/turquoise)
    options = {"colors": {"MOUNTAIN": "linear-gradient(90deg, #aa9cfc, #fc9ce7)"}}
    
    # Render the result
    displacy.render(ex, style="ent", manual=True, jupyter=True, options=options)

# =====================================================================
# TESTING ON VARIOUS EXAMPLES
# =====================================================================

print("Recognition results:")
print("-" * 50)

visualize_ner("Yesterday we decided to travel to Mount Everest, it was awesome.")
visualize_ner("The highest peak in Ukraine is Hoverla, but Kilimanjaro is much higher.")
visualize_ner("I want to climb Mont Blanc next summer with my friends.")
```
